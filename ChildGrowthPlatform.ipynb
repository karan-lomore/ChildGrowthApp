<h1><center>Child Growth Assessment Platform</center></h1>

This prototype classify growth for children age younger than 18 years old. The calculation is based on Indian Academy of Pediatrics (IAP) and World Health Organization (WHO) growth standards. The Percentiles are calculated for the following parameters.

<b> Under Five years old children </b>
<ol>
<li>Weight for Age </li>
<li>Height for Age</li>
<li>Head Circumference for Age</li>
<li>Arm Circumference for Age</li>
<li>Weight for Height/Length</li>
<li>Body Mass Index (BMI) for Age</li>
</ol>

<b> Above Five years old children </b>
<ol>
<li>Weight for Age </li>
<li>Height for Age</li>
<li>Body Mass Index (BMI) for Age</li>
</ol>

```python
%matplotlib widget
import ipywidgets as ipy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from IPython.display import display, Markdown, clear_output, HTML
from datetime import date
from ipywidgets import HBox, Label, Layout
from datetime import datetime
```

```python
def calculate_age_func(age_days):
    
    age_yrs = age_days / 365
    
    if age_yrs <= 5:
        types = 'below five'
    else:
        types = 'above five'
    
    return types


def dob_to_age_func(dob_ddmmyyyy):
    
    crnt_date = datetime.now()
    
    dob_format_chnge = datetime.strptime(dob_ddmmyyyy,'%Y-%m-%d')
    
    age_in_days = crnt_date - dob_format_chnge
    
    age_in_yrs = age_in_days.total_seconds() / (365.242 * 24 * 3600)
    
    age_in_yrs = round(age_in_yrs, 2)
    
    age_in_yrs_int = int(age_in_yrs)
    
    age_remaining_months = (age_in_yrs - age_in_yrs_int) * 12
    
    age_remaining_months_int = int(age_remaining_months)
    
    age_yrs_months_list = [age_in_yrs_int, age_remaining_months_int]
    
    return age_in_yrs_int, age_in_yrs, age_yrs_months_list


def get_iap_pctl_func(guidelines, gender, age):
    
    if guidelines == 'IAP':
        BMI = True
        fat = True
        age_ym = age
        if gender =='Male':
            weight_data = pd.read_csv('IAP-Dataset/male_weight_SD.csv')
            height_data = pd.read_csv('IAP-Dataset/male_height_SD.csv')

        elif gender =='Female':
            weight_data = pd.read_csv('IAP-Dataset/female_weight_SD.csv')
            height_data = pd.read_csv('IAP-Dataset/female_height_SD.csv')
        
        rownumber_weight = weight_data[weight_data['Age'] == age_ym].index[0]
        weight_data.columns = ['Age','3rd','10th','25th','50th','75th','90th','95th']
        wt_coln = weight_data.columns
        
        rownumber_height = height_data[height_data['Age'] == age_ym].index[0]
        height_data.columns = ['Age', '3rd', '10th', '25th', '50th', '75th', '90th', '95th']
        ht_coln = height_data.columns
    
    return weight_data, wt_coln, height_data, ht_coln


def plotting(data_list, columns_list, parameter_list, variable_list, age_ym, guidelines):
    
    plt.close()
    
    f, (a0, a1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [3, 3]}, figsize=(8, 4))
    
    if parameter_list[0] == 'Weight':
        data = data_list[0]
        columns_list = columns_list[0]
        parameter = parameter_list[0]
        variable = variable_list[0]
        #a0.plot(data.iloc[:,0].values,data['3rd'], linestyle='-', color = '#cf3933')
        a0.plot(data.iloc[:,0],data['10th'], linestyle='-', color = '#cf3933',label='Below 10th percentile')
        a0.plot(data.iloc[:,0],data['25th'], linestyle='-', color = '#7f7f7f',label='10th-25th percentile')
        a0.plot(data.iloc[:,0],data['50th'], linestyle='-', color = '#468d50',label='25th-50th percentile')
        a0.plot(data.iloc[:,0],data['75th'], linestyle='-', color = '#7f7f7f',label='50-75th percentile')
        a0.plot(data.iloc[:,0],data['90th'], linestyle='-', color = '#cf3933',label='75th-90th percentile')
        a0.plot(data.iloc[:,0],data['95th'], linestyle='-', color = '#cf3933',label='Above 90th percentile')
        
        a0.stackplot(data.iloc[:,0].values,data['95th'], color = '#efbdbd')
        a0.stackplot(data.iloc[:,0].values,data['90th'], color = '#f9e7b2')
        a0.stackplot(data.iloc[:,0].values,data['75th'],color = '#c5dcc7') 
        a0.stackplot(data.iloc[:,0].values,data['25th'],color = '#f9e7b2') 
        a0.stackplot(data.iloc[:,0].values,data['10th'],color = '#efbdbd') 
        #a0.stackplot(data.iloc[:,0].values,data['3rd'],color = '#efbdbd')
        
        if 'CDC' in guidelines:
            x = 'Age (in months)'
            y = 4
        elif 'IAP' in guidelines:
            x = 'Age'
            y = 4
        else:
            x = 'Month'
            y = 48
            
        a0.scatter(age_ym, variable, color = '#42a1f5', s = 30)
        #plt.text(age_ym+y+0.2, data.loc[data[x] == age_ym+y, '3rd'], '3rd', fontsize=12, horizontalalignment='center',verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '10th'], '10th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '25th'], '25th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '50th'], '50th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '75th'], '75th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '90th'], '90th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a0.text(age_ym+y+0.3, data.loc[data[x] == age_ym+y, '95th'], '95th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')    
        
        if age_ym >= 9:
            a0.set_xlim(age_ym-y,age_ym+y)
            a0.set_ylim(data.loc[data[x] == age_ym-y, '10th'].to_list()[0], data.loc[data[x] == age_ym+y, '95th'].to_list()[0])
        else:
            a0.set_xlim(5, age_ym+y)
            a0.set_ylim(data.loc[data[x] == age_ym, '10th'].to_list()[0], data.loc[data[x] == age_ym+y, '95th'].to_list()[0])
        
        a0.set_xlabel('{}'.format('Age (years)'), fontsize=10)
        a0.set_ylabel('{}'.format('Weight (kg.)'), fontsize=10)
        a0.xaxis.set_tick_params(labelsize=8)
        a0.yaxis.set_tick_params(labelsize=8)
        
    
    if parameter_list[1] =='Height':
        data = data_list[1]
        columns_list = columns_list[1]
        parameter = parameter_list[1]
        variable = variable_list[1]
        #a1.plot(data.iloc[:,0],data['3rd'], linestyle='-', color = '#cf3933')
        a1.plot(data.iloc[:,0],data['10th'], linestyle='-', color = '#e2c644',label='Below 10th percentile')
        a1.plot(data.iloc[:,0],data['25th'], linestyle='-', color = '#dcdcdc',label='10th-25th percentile')
        a1.plot(data.iloc[:,0],data['50th'], linestyle='-', color = '#bfbfbf',label='25th-50th percentile')
        a1.plot(data.iloc[:,0],data['75th'], linestyle='-', color = '#dcdcdc',label='50-75th percentile')
        a1.plot(data.iloc[:,0],data['90th'], linestyle='-', color = '#dcdcdc',label='75th-90th percentile')
        a1.plot(data.iloc[:,0],data['95th'], linestyle='-', color = '#dcdcdc',label='Above 90th percentile')
        a1.stackplot(data.iloc[:,0].values,data['95th'], color = '#c5dcc7')
        a1.stackplot(data.iloc[:,0].values,data['90th'], color = '#c5dcc7')
        a1.stackplot(data.iloc[:,0].values,data['75th'],color = '#c5dcc7') 
        a1.stackplot(data.iloc[:,0].values,data['25th'],color = '#f9e7b2') 
        a1.stackplot(data.iloc[:,0].values,data['10th'],color = '#efbdbd') 
        #a1.stackplot(data.iloc[:,0].values,data['3rd'],color = '#efbdbd')
        
        if 'CDC' in guidelines:
            x = 'Age (in months)'
            y = 4
        elif 'IAP' in guidelines:
            x = 'Age'
            y = 4
        else:
            x = 'Month'
            y = 48
            
        a1.scatter(age_ym, variable, color = '#42a1f5',s = 30)
        #plt.text(age_ym+y+0.2, data.loc[data[x] == age_ym+y, '3rd'], '3rd', fontsize=12, horizontalalignment='center',verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '10th'], '10th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '25th'], '25th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '50th'], '50th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '75th'], '75th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '90th'], '90th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')
        a1.text(age_ym+y+0.4, data.loc[data[x] == age_ym+y, '95th'], '95th', fontsize=8, horizontalalignment='center',
        verticalalignment='center')    
        
        if age_ym >= 9:
            a1.set_xlim(age_ym-y,age_ym+y)
            a1.set_ylim(data.loc[data[x] == age_ym-y, '10th'].to_list()[0], data.loc[data[x] == age_ym+y, '95th'].to_list()[0])
        else:
            a1.set_xlim(5,age_ym+y)
            a1.set_ylim(data.loc[data[x] == age_ym, '10th'].to_list()[0], data.loc[data[x] == age_ym+y, '95th'].to_list()[0])
        
        a1.set_xlabel('{}'.format('Age (years)'), fontsize=10)
        a1.set_ylabel('{}'.format('Height (cm.)'), fontsize=10)
        a1.xaxis.set_tick_params(labelsize=8)
        a1.yaxis.set_tick_params(labelsize=8)
        
        f.suptitle('Child Growth Assessment') 
        f.tight_layout(pad=2.0)
        plt.show()
    
    return

widgets = {}

widgets['grid'] = ipy.GridspecLayout(7, 2)

widgets['DOB'] = ipy.DatePicker(
           description='Date of birth', disabled = False)

widgets['DOV'] = ipy.DatePicker(
           description='Date of visit', disabled = True)

widgets['DOV'].value = date.today()

widgets['sex'] = ipy.Dropdown(
       options=['Male', 'Female'],
       value='Male',
       description='Sex:')

widgets['growthchart'] = ipy.Dropdown(
       options=['IAP', 'WHO'],
       value='IAP',
       description='Chart:')

widgets['parameter'] = ipy.Dropdown(
       options=['Z Scores', 'Percentiles'],
       value='Percentiles',
       description='Parameter:')

widgets['weight'] = ipy.BoundedFloatText(description='Weight (kg.)', value=0, min=0, max=80, step=.5)

widgets['height'] = ipy.BoundedFloatText(description='Height (cm.)', value=0, min=0, max=170, step=.5)

widgets['head_circum'] = ipy.BoundedFloatText(description='Head Circumference (cm.)', value=0, min=0, max=170, step=.5)

widgets['arm_circum'] = ipy.BoundedFloatText(description='Arm Circumference (cm.)', value=0, min=0, max=170, step=.5)

widgets['grid'][0, 0] = widgets['DOB']
widgets['grid'][1, 0] = widgets['DOV']
widgets['grid'][2, 0] = widgets['sex']
widgets['grid'][3, 0] = widgets['growthchart']
widgets['grid'][4, 0] = widgets['parameter']

next_button = ipy.Button(description='Next')
next_box_layout = ipy.Layout(display='flex',
                flex_flow='column',
                align_items='center',
                width='92%')

next_box = ipy.HBox(children=[next_button], layout=next_box_layout)
widgets['grid'][6, 0] = next_box
next_out = ipy.Output()

def next_button_clicked(_):
      with next_out:
            clear_output()
            dob = widgets['DOB'].value
            dov = widgets['DOV'].value
            age_days = (dov - dob).days
            child_cat = calculate_age_func(age_days)
            
            if child_cat == 'below five':
                widgets['grid'][0, 1] = widgets['weight']
                widgets['grid'][1, 1] = widgets['height']
                widgets['grid'][2, 1] = widgets['head_circum']
                widgets['grid'][3, 1] = widgets['arm_circum']
                
            else:
                widgets['grid'][0, 1] = widgets['weight']
                widgets['grid'][1, 1] = widgets['height']


next_button.on_click(next_button_clicked)

submit_button = ipy.Button(description='Submit')
submit_box_layout = ipy.Layout(display='flex',
                flex_flow='column',
                align_items='center',
                width='92%')
submit_box = ipy.HBox(children=[submit_button], layout=submit_box_layout)

widgets['grid'][6, 1] = submit_box

submit_out = ipy.Output()

def submit_button_clicked(_):
    with submit_out:
        clear_output()
        wt = widgets['weight'].value
        ht = widgets['height'].value
        dob = widgets['DOB'].value
        dov = widgets['DOV'].value
        sex = widgets['sex'].value
        growth_chart = widgets['growthchart'].value
        chart_type = widgets['parameter'].value
        
        age_in_yrs_int, age_in_yrs, age_yrs_months_list = dob_to_age_func(str(widgets['DOB'].value))
        
        age = age_yrs_months_list[0]
        actual_months = age_yrs_months_list[1]
        
        if actual_months <4:
            months = 0
        elif 4 <= actual_months < 9:
            months = 6
        else:
            age+=1
            months = 0
        
        age_yrs = age + (months/12)
        
        weight_data, wt_coln, height_data, ht_coln = get_iap_pctl_func(growth_chart, sex, age_yrs)  
        
        return display(plotting([weight_data, height_data], [wt_coln, ht_coln], ['Weight', 'Height'], [wt, ht], age_yrs, growth_chart))
        
display(widgets['grid'])

submit_button.on_click(submit_button_clicked)
ipy.VBox([submit_out])
```
